```python
from bs4 import BeautifulSoup
import requests
```

```python
import pandas as pd
```

```python
data=pd.read_excel("input.xlsx")
```

```python
y=0
title_df=[]
desc_df=[]
nos=[]
for x in data.URL:
    y=y+1
    print(y)
    html_text=requests.get(x).text
    soup=BeautifulSoup(html_text,'lxml')
    title=soup.find('h1',)
    if title==None:
        nos.append(y)
        continue
    else:
        title=soup.find('h1',).text
    desc=soup.find('div',class_='td-post-content tagdiv-type')
    if desc==None:
        desc=soup.find('div',class_='tdb-block-inner td-fix-index').text.replace('\n','')
    else:
        desc=soup.find('div',class_='td-post-content tagdiv-type').text.replace('\n','')
    pre=soup.find('pre',class_='wp-block-preformatted')
    if pre==None:
        desc_final=desc
    else:
        pre=soup.find('pre',class_='wp-block-preformatted').text
        desc_final=desc.replace(pre,'')
    title_df.append(title)
    desc_df.append(desc_final)
```

```python
dict={'title':title_df,'description':desc_df}
df=pd.DataFrame(dict)
df.to_csv('article.csv')
df1=pd.read_csv('article.csv')
df1.head()
```
